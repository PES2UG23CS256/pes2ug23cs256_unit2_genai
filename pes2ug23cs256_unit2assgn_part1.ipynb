```python
from dotenv import load_dotenv
load_dotenv()

import getpass
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your Google API Key: ")
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.0)
```

```python
template = ChatPromptTemplate.from_messages([
    ("system",
     "You are a movie expert. When given a movie name, state its release year "
     "and calculate how many years ago that was from the current year 2025. "
     "Be concise and clear."),
    ("human", "Movie: {movie}")
])
```

```python
parser = StrOutputParser()
```

```python
chain = template | llm | parser
```

```python
result = chain.invoke({"movie": "Inception"})
print(result)
```

```python
result = chain.invoke({"movie": "The Godfather"})
print(result)
```

```python
movie_name = input("Enter a movie name: ")
result = chain.invoke({"movie": movie_name})
print(result)
```
